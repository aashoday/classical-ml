```python
# Import libraries
import numpy as np
import  matplotlib.pyplot as plt
np.__version__
```

```python
# 1. Generate a synthetic dataset
 
# Let the hypothesis be y = 3x + 7
weight = 3
bias = 7

# Create data 
start = -10
end = 10
step = 1

X = np.arange(start, end, step) # Features
y = weight * X + bias # Labels

print(X, len(X))
print(y, len(y))
```

```python
# 2. Create a 80-20 training testing split

# Set a splitting point
split_point = int(0.8 * len(X)) # 80% training data and 20%testing data

# Create training and testing splits
X_train, y_train = X[:split_point], y[:split_point]
X_test, y_test = X[split_point:], y[split_point:]

# Print length
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
# 3. Plot current data

def plot_predictions(train_data=X_train,
                    train_labels=y_train,
                    test_data=X_test,
                    test_labels=y_test,
                    predictions=None):
    """
    Plot's training data, testing data and comparest predictions.
    """
    plt.figure(figsize=(16,9))

    # Plot training data 
    plt.scatter(train_data, train_labels, c="b" , s=4, label="Training Data")

    # Plot testing data
    plt.scatter(test_data, test_labels, c="r" , s=4, label="Testing Data")

    # Plot prediction if exist
    if predictions is not None:
        plt.scatter(test_data, predictions, c="g", s=4, label="Predicitions")

    # Show legend
    plt.legend(prop={"size": 14})
```

```python
plot_predictions()
```
